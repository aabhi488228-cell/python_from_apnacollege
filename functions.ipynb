```python
#print average of 3 numbers
def calc_avg(a,b,c):
    sum = a+b+c
    avg = sum/3
    print(avg)
calc_avg(1,2,4)
```

```python
def calc_sgpa(a,b,c,d,e,f,g,h):
    sum = a+b+c+d+e+f+g+h
    avg = sum/8
    print(avg)
calc_sgpa(10,10,10,9,8,8,8,8)
```
