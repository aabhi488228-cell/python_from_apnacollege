```python
#pracrice02:search a number x in this tuple using loop
num = (1,4,9,16,25,36,49,64,81,100)
x = int(input("enter the number x:"))
idx = 0
for el in num:
    if (el == x):
        print("the element x is found at idx", idx)
        break
    idx += 1
else:
    print("the element x is not found")
        
```
